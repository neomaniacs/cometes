### Clean dataframe

```python
# process tel
def tel_p(val): 
    if(not val):
        return ''
    val=str(val)
    if(len(val)==11):
        return '0'+val
    elif(len(val)==9):
        return '0'+val
    else:
        return '+'+val

debug=0
headers = ['time', 'email', 'ecole', 'nom_enfant', 'prenom_enfant', 'classe_septembre', 'nom_pere', 'nom_mere'
           ,'tel_pere','tel_mere', 'adresse_post', 'email_corresp', 'email2', 'photo', 'sortie_seule'
           , 'ps', 'ms', 'gs', 'cp', 'ce1', 'ce2', 'cm1', 'cm2']
# mapping avec en-têtes
df.columns=headers

# drop all rows without name, last name and school of children
df.dropna(subset=['ecole', 'nom_enfant', 'prenom_enfant'], inplace=True)

# format time
df['time']=df['time'].apply(lambda x: pd.datetime.strptime(x[:18], '%Y/%m/%d %H:%M:%S'))

# format email in lowercase
df['email']= df['email'].str.lower()
df['email_corresp']= df['email_corresp'].str.lower()
df['email2']= df['email2'].str.lower()

# format tel
# --> int to keep all numbers
#df['tel_pere']=df['tel_pere'].astype('int')
df['tel_pere'] = df['tel_pere'].replace(np.nan, '', regex=True)  
df['tel_mere'] = df['tel_mere'].replace(np.nan, '', regex=True)       
       
# transform to string
df['tel_pere']=df['tel_pere'].astype('str')
# --> to str to add 0 or +
df['tel_pere']=df['tel_pere'].apply(lambda x: tel_p(x))

# test tel_mere    
df['tel_mere']=df['tel_mere'].astype('str')
df['tel_mere'].replace('\\.0', '', regex=True)
df['tel_mere']=df['tel_mere'].apply(lambda x: tel_p(x))

# last name        
df['nom_enfant'] = df['nom_enfant'].str.upper()
df['nom_enfant'].head()
## ecole
#df.loc[df['ecole'].str.contains('Marthe') == True, 'ecole'] = 'sainte_marthe'
#df.loc[df['ecole'].str.contains('Coeur') == True, 'ecole'] = 'sacre_coeur'
## class
df.loc[df['classe_septembre'].str.contains('Petite') == True, 'classe_septembre'] = 'PS'
df.loc[df['classe_septembre'].str.contains('Moyenne') == True, 'classe_septembre'] = 'MS'
df.loc[df['classe_septembre'].str.contains('Grande') == True, 'classe_septembre'] = 'GS'
df['classe_septembre'].str.lower()

## sortie et photo
df['photo'] = df['photo'].str.lower()
df['sortie_seule'] = df['sortie_seule'].str.lower()

## format 
df['prenom_enfant'] = df['prenom_enfant'].str[0:1].str.upper()+df['prenom_enfant'].str[1:].str.lower()
    
# remove into address    
#df['adresse_post'] = df['adresse_post'].map(lambda x: x.lstrip(' ').rstrip('\n'))    
df=df.replace('\n', ' ', regex=True) 
    
# remove empty email
df=df[df['email'] != "@"]    

# remove
df = df[df['classe_septembre'].notna()]

# delete if more than 6 NAN
df=df.dropna(thresh=6)

# delete duplicate based on 
df=df.drop_duplicates(subset=['email', 'ecole', 'nom_enfant', 'prenom_enfant', 'ps', 'ms', 'gs', 'cp', 'ce1', 'ce2', 'cm1', 'cm2'], keep="last")

# drop file
# delete an old "output" file 
try:
    os.remove('files/output.csv')
except:
    print('the output file doesn\'t exist')
```

### Excec data processing

```python
test=1
debug=1
exect=1
match_activite = '(ACT\d{2,4})\s?\|\s?(.*)\s?\|\s?(.*)\s?\|\s(\d{2,3})\s€'
# Parse file in timestamp order
for index,row in df.iterrows():
    print('//-------------------------------')    
    #print ('row : ',row)       
    print('classe septembre : ',row['classe_septembre'])
    # check if row exists
    if(row.isnull == True):
        if debug>1:
            print('--> continue because row is null')
            print('Row: ', row)
        continue
    # test classe septembre
    elif(pd.isnull(row['classe_septembre']) == True):
        if debug>1:
            print('--> continue because classe septemre is null')
            print(row['classe_septembre'])
        continue
        
    classe = row['classe_septembre'].lower()
    if debug>1:
       print('Act : ',row[classe])
    
    # test value for that classe    
    if(pd.isnull(row[classe]) == True):
        if debug>1:
            print('--> continue because row/classe is null')
        continue
    
    # check format activities
    if(row[classe].find(' - ') != -1):
        print('Check activities: ',row[classe])
    
    # if several activities
    if(row[classe].find(';') != -1):
        activities = row[classe].split(';')
        nbre=len(activities)
        if debug==1:
            print('--> Nbre activites --> ',nbre)
        for i in range(0,nbre):
            act=activities[i].strip()
            if debug==1:
                    print ('Multi/activite_', i,' --> ',act)
                    print('Index :',index)
                    m=re.search(match_activite, act)
                    if(pd.isnull(m) == True):
                        print('--> no match because m is null 1')
                        continue
                    print('1/ Match(0) :',m.group(0))
                    print('1/ Match(1) :',m.group(1))
                    print('1/ Match(2) :',m.group(2))
                    print('1/ Match(3) :',m.group(3))
                    print('1/ Match(4) :',m.group(4))
            df['activite_'+str(i+1)][index]=act
            df['code_'+str(i+1)][index]=m.group(1)
            code=m.group(1)
            print('regexp code : ', code)
            df['label_'+str(i+1)][index]=m.group(2)
            df['location_'+str(i+1)][index]=m.group(3)
            df['price_'+str(i+1)][index]=m.group(4)
            #index=str(i+1)
            index_act=str(i+1)
            
            # check quota
            print('Mapped code : ',code, ' quota left : ',locals()[code])

            if(locals()[code]>0):
                print('code mapped')
                locals()[code] -=1
                print('--> new quota :', locals()[code])
                print('Confirmed for act :', str(i+1))
                df['total_act'][index]+=int(m.group(4))
                df['confirmed_'+str(i+1)][index]=1
                df['status_'+str(i+1)][index]='Confirmé'
            else:
                df['confirmed_'+str(i+1)][index]=0
                df['status_'+str(i+1)][index]='En attente'
                print('////// ----> quota excessed')
    else:
        act=row[classe].strip()
        if debug==1:
            print('Mono/ 1 seule activite / activite_1, on act : ',act)
            print('Index :',index)
        if debug==1:
            print('activite : ', act)
            m=re.search(match_activite,act)
            if(pd.isnull(m) == True):
                print('--> no match because m is null 2')
                continue
            print('2/ Match(0) :',m.group(0))
            print('2/ Match(1) :',m.group(1))
            print('2/ Match(2) :',m.group(2))
            print('2/ Match(3) :',m.group(3))
            print('2/ Match(4) :',m.group(4))
        df['activite_1'][index]=act
        df['code_1'][index]=m.group(1)
        code=m.group(1)
        print('regexp code : ', code)
        df['label_1'][index]=m.group(2)
        df['location_1'][index]=m.group(3)
        df['price_1'][index]=m.group(4)
        index_act = 1
        test+=1
        
        # check quota
        print('Mapped code : ',code, ' quota left : ',locals()[code])

        if(locals()[code]>0):
            print('code mapped')
            locals()[code] -=1
            print('--> new quota :', locals()[code])
            print('Confirmed for act :', str(index_act))
            # if confirmed, add total price
            df['total_act'][index]+=int(m.group(4))
            df['confirmed_'+str(index_act)][index]=1
            df['status_'+str(index_act)][index]="Confirmé"
        else:
            df['confirmed_'+str(index_act)][index]=0
            df['status_'+str(index_act)][index]="En attente"
            print('////// ----> quota excessed')

# sort by time
df=df.sort_values(by=['time'])
        
if debug==1:
    print('-------------------------------//')
if exect==1:
    df.to_csv('files/output.csv', index=False)
```
